<h2>Chi Square Test</h2>

<h3>What is Chi-Square Test?</h3>
<p>There are two types of chi-square tests. Both use the chi-square statistic and distribution for different purposes:</p>
<ul>
  <li>A chi-square goodness of fit test determines if a sample data matches a population.</li>
  <li>A chi-square test for independence compares two variables in a contingency table to see if they are related. In a more general sense, it tests to see whether distributions of categorical variables differ from each other.</li>
</ul>
<p>A very small chi square test statistic means that your observed data fits your expected data extremely well. In other words, there is a relationship.</p>
<p>A very large chi square test statistic means that the data does not fit very well. In other words, there isn’t a relationship.</p>
<p>In data mining, mostly a chi-square statistic is one way to show a relationship between two categorical variables. In statistics, there are two types of variables: numerical (countable) variables and non-numerical (categorical) variables. The chi-squared statistic is a single number that tells you how much difference exists between your observed counts and the counts you would expect if there were no relationship at all in the population.</p>

<h3>Formula for Chi-Square Statistic</h3>
<p>The formula for the chi-square statistic used in the chi square test is:</p>
<pre>
  χ² = Σ[(O - E)² / E]
  Where:
    O = Observed frequency
    E = Expected frequency
</pre>

<p>A low value for chi-square means there is a high correlation between your two sets of data. In theory, if your observed and expected values were equal ("no difference") then chi-square would be zero — an event that is unlikely to happen in real life.</p>

<h3>Chi-Square Table</h3>
<p>The chi square table below is used in hypothesis testing. It helps you to decide whether to accept or reject a hypothesis. A hypothesis is our supposition. This hypothesis basically gives us the idea whether our observed values are close to our expectation or not. We normally have two types of hypothesis. These types are given below:</p>
<ul>
  <li>Null Hypothesis - denoted as H0</li>
  <li>Alternate Hypothesis - denoted as H1</li>
</ul>

```python
import numpy as np  # import the library
from scipy.stats import chi2_contingency   # import the library of python to calculte the chi-square

# Observed data (from Example 1)
# Data set
observed_value = np.array([
    [25, 31],  # Dell Latitude
    [20, 18],  # HP Envy
    [18, 10],  # Mac Book Air
    [20, 20],  # Oppo F9 Pro
    [15, 8],   # Apple iPhone 7
    [25, 28],  # Apple iPhone X
    [22, 34],  # Samsung Galaxy S10
])

# Perform Chi-Square test
chi2, p, dof, expected = chi2_contingency(observed_value) # chi2_contigency is method to calculte the chi-square, chi2 store value of chi-square, p store P-value
# dof store the degree of freedom, and excepted store the expected frequency

# Display the results
print(f"Chi-Square Value: {chi2}")
print(f"P-Value: {p}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies: \n{expected}")

# Interpretation of the result
alpha = 0.05  # significance level
if p < alpha:
    print("Reject the null hypothesis: Sale is not according to expectations.")
else:
    print("Fail to reject the null hypothesis: Sale is according to expectations.")
```
