```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
# import the data from the csv file into dataframe
df = pd.read_csv('D:\\Raza data\\Data Mining\\Corelation_and_chi_square\\Ice Cream Sales - temperatures.csv')
print(df.shape)   # How to know about the shape the dataframe
print(df.columns)
t=type(df.iloc[0])
print(t)
df
# Calculate Pearson correlation coefficient with library
correlation, _ = pearsonr(df['Temperature'], df['Ice Cream Profits'])
print("corelation of dataset" , correlation)

```

```python
# To plot the graph of that correlation
plt.scatter(df['Temperature'], df['Ice Cream Profits'], color='red')

# Set the title and labels for the axes
plt.title('Correlation Between Temp and Ice cream')
plt.xlabel('Temperature')
plt.ylabel('Ice Cream Profits')

# Show the plot
plt.show()
```
